```python
relevant_inflation = pd.read_csv("inflation.csv").tail(35)

relevant_inflation.tail()
```

```python
def calc_inf(year1, year2):
    inf_factor = 1
    for r in relevant_inflation.iterrows():
        if r[1]['Year'] >= year1 and r[1]['Year'] <= year2:
            inf_factor *= (1 + 0.01*float(r[1]['can_inflation'][:-2]))
    return inf_factor

calc_inf(1985, 2017)
```

```python
rankp = pd.read_csv("ranked_provinces.csv")

def get_expenditure(name):
    for r in expenditure_by_year[expenditure_by_year['Year'] == 2018].iterrows():
        return r[1][3:][province_full.index(name)]
    
rankp['Expenditure (millions)'] = [get_expenditure(x) for x in rankp['GEO']]

rankp['CBC rank'] = [1, 3, 2, 4, 6, 5, 9, 10, 8 , 7]
rankp['Frontier rank'] = [2, 8, 1, 4, 5, 6, 9, 10, 7, 3]

rankp
```
