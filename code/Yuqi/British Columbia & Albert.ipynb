```python
relevant_inflation = pd.read_csv("inflation.csv").tail(35)
```

```python
def calc_inf(year1, year2):
   inf_factor = 1
   for r in relevant_inflation.iterrows():
       if r[1]['Year'] >= year1 and r[1]['Year'] <= year2:
           inf_factor *= (1 + 0.01*float(r[1]['can_inflation'][:-2]))
   return inf_factor
```
